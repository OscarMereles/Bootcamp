```python
class Animales:
    especie= "ave"

    def __init__(self,nombre,color):
        self.nombre = nombre
        self.color = color

    def volar(self,firma):
        print("I BELIEVE I CAN FLY, I BELIEVE I CAN..."+firma)

    def cantar(self,firma):
        print("Pio Pio, me pintaron pajaritos en el aire "+firma)

gorrion = Animales("Ricardo","Rojo")
carpintero = Animales("Loquillo","Azul")
print("Holi, soy un "+gorrion.especie+", me llamo "+gorrion.nombre+" y soy de color "+gorrion.color)
gorrion.volar(gorrion.nombre)
gorrion.cantar(gorrion.nombre)
print("Holi, soy un "+carpintero.especie+", me llamo "+carpintero.nombre+" y soy de color "+carpintero.color)
carpintero.volar(carpintero.nombre)
carpintero.cantar(carpintero.nombre)
```

```python
class Ave:
    Alas=True

    def __init__(self,nombre, vuela):
        self.nombre= nombre
        self.vuela= vuela


class Pinguino(Ave):
    color="blanco y negro"

pengu= Pinguino("SuperPengu","puedo volar")

print(f"Hola soy {pengu.nombre} y {pengu.vuela}")
```

```python
class Hereo:
    moralidad = "justiciero"

    def __init__(self,nombre,poder):
        self.nombre = nombre
        self.poder = poder

    def atacar(self, objetivo,ataque):
        print(f"el heroe uso su {ataque} contra {objetivo}")

class Villano:
    moralidad = "Bandido"

    def __init__(self,nombre,poder):
        self.nombre = nombre
        self.poder = poder

    def atacar(self, objetivo,ataque):
        print(f"el villano uso su {ataque} contra {objetivo}")

flash= Hereo("Flash","supervelocidad")
print(f"el {flash.moralidad}, {flash.nombre} usa su {flash.poder}")
flash.atacar("Gorilla Grodd","golpe supersonico")
grodd= Villano("Gorilla Grodd", "poder mental")
print(f"el {grodd.moralidad}, {grodd.nombre} usa su {grodd.poder}")
grodd.atacar("Flash","ataque de alucinaciones")
```
