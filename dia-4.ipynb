```python
diccionario={}                                          #diccionario vacio

diccionario["nombre"]=input("Ingrese su nombre ")       #se guarda la entrada de datos en la clave nombre
print(diccionario)                                      #se imprime el contenido

diccionario["apellido"]=input("Ingrese su apellido ")   #y asi sucesivamente...
print(diccionario)

diccionario["edad"]=input("Ingrese su edad ")
print(diccionario)

diccionario["altura"]=input("Ingrese su altura ")
print(diccionario)

diccionario["ciudad"]=input("Ingrese su ciudad ")
print(diccionario)
```

```python
datos={}

def menu_tarjeta(orden, diccionario):
    if orden=="crear":
        nombre_clave=input("Ingrese que tipo de dato es ")
        valor_dato=input("Ingrese el valor del dato")
        diccionario[nombre_clave]=valor_dato
    elif orden=="modificar":
        print(diccionario.keys())
        eleccion_mod=input("Elija que dato desea modificar: ")
        valor_dato=input("Ingrese el valor del dato")
        diccionario[eleccion_mod]=valor_dato
    elif orden=="eliminar":
        print(diccionario.keys())
        eleccion_eli=input("Elija que dato desea eliminar: ")
        del diccionario[eleccion_eli]
    return diccionario

while True:
    print(datos)
    instruccion=input("Indique que desea hacer: 'crear' - 'modificar' - 'eliminar' datos o ingrese 0 para salir")
    print(instruccion)
    while instruccion != "crear" and instruccion != "modificar" and instruccion != "eliminar":
        if instruccion =="0":
            break
        instruccion=input("Comando incorrecto, indique que desea hacer: 'Crear' - 'Modificar' - 'Eliminar' datos o ingrese 0 para salir")
    print(menu_tarjeta(instruccion,datos))
```

```python
contactos={}

def menu_contacto(orden, diccionario):
    if orden=="crear":
        nombre_contacto=input("Ingrese nombre del contacto ")
        numero_telefono=input("Ingrese el numero de contacto ")
        correo=input("Ingrese correo electronico del contacto ")
        diccionario[nombre_contacto]={"numero":numero_telefono,"correo":correo}
    elif orden=="modificar":
        print(diccionario.keys())
        eleccion_mod=input("Elija que contacto que desea modificar: ")
        valor_dato=input("Ingresa la  clave a modificar ")
        diccionario[eleccion_mod]=valor_dato
    elif orden=="mostrar":
        print(diccionario.keys())
        eleccion_mostrar=input("Elija que dato desea eliminar: ")
        print(diccionario.items(eleccion_mostrar))
    return diccionario

while True:
    print(contactos)
    instruccion=input("Indique que desea hacer: 'crear' - 'modificar' - 'mostrar' contactos o ingrese 0 para salir")
    print(instruccion)
    while instruccion != "crear" and instruccion != "modificar" and instruccion != "mostrar":
        if instruccion =="0":
            break
        instruccion=input("Comando incorrecto, indique que desea hacer: 'crear' - 'modificar' - 'mostrar' contactos o ingrese 0 para salir")
    print(menu_contacto(instruccion,contactos))
```
