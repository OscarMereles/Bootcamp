```python
caracter1='a'                               #Variable que almacena el primer caracter o letra
caracter2='b'                               #Varaible que almacena el segundo caracter o letra
def comparar_letras(letra1,letra2):         #Funcion que compara las letras y recibe los parametros
    if letra1==letra2:                      #Condicional que analiza la igualdad entre caracteres
        print("Las letras son iguales")     #Salida por pantalla que afirma igualdad entre letras
    else:
        print("Las letras son diferentes")  #Salida por pantalla que niega la igualdad entre letras
comparar_letras(caracter1,caracter2)        #Llamado a la funcion que compara las letras y se le pasan los argumentos
```

```python
def contar_letras(palabra,numero):                              
    if len(palabra)== numero:                                   
        return True
    else:
        if palabra=="0" and numero==0:
            return "break"
        return False

while True:
    word=input("ingrese palabra:")
    nro=int(input("ingrese cuantas letras tiene tu palabra: "))
    if contar_letras(word,nro)!="break":
        print(contar_letras(word,nro))
    else:
        break
```
