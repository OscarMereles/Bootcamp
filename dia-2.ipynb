```python
frase=input("ingrese una frase")
contador=0
contador_min=0
contador_may=0
for i in range(len(frase)):
    if frase[i]=='o' or frase[i]=='O':
        if frase[i]=='o':
            contador_min+=1
        if frase[i]=='O':
            contador_may+=1
        print("O/o encontrado")
        contador+=1
    else:
        print(frase[i])
print("Se encontraron: ",contador," O's")
print("Se encontraron: ",contador_min," o's minusculas")
print("Se encontraron: ",contador_may," O's mayusculas")
```
